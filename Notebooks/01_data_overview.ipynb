```python
df['vote_count'] = pd.to_numeric(df['vote_count'], errors='coerce')

print(f"Minimum vote_count: {df['vote_count'].min()}")
print(f"Maximum vote_count: {df['vote_count'].max()}")
print(f"Average vote_count: {df['vote_count'].mean():.2f}")

print(f"Movies with vote_count < 5: {df[df['vote_count'] < 5].shape[0]}")
print(f"Movies with vote_count < 10: {df[df['vote_count'] < 10].shape[0]}")
```

```python
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')

low_vote_hight_rating = df[(df['vote_count'] < 5) & (df['vote_average'] >= 8)]

print(f"Low vote count but high rating: {low_vote_hight_rating.shape[0]}")
```

```python
df['overview'] = df['overview'].fillna('')

df['overview_length'] = df['overview'].apply(len)

print(f"Minimum overview length: {df['overview_length'].min()}")
print(f"Maximun overview length: {df['overview_length'].max()}")
print(f"Average overview length: {df['overview_length'].mean():.2f}")

print(f"Very short overvies < 10 characters: {df[df['overview_length'] > 10].shape[0]}")
```
