```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("../Data/processed/cleaned_dataset.csv")
df.head()
```

```python
tfidf = TfidfVectorizer(
    max_features=20000
)
tfidf_matrix = tfidf.fit_transform(df['description'])

tfidf_matrix.shape
```

```python
indices = pd.Series(df.index, index=df['original_title']).to_dict()
```

```python
indices
```

```python
titles = ["jumanji"] 
```

```python
if isinstance(titles, str):
    titles = [titles]
print(titles)
```

```python
valid_indices = [indices[t] for t in titles if t in indices]
```

```python
if valid_indices:
    # similarity compute
    sim_matrix = cosine_similarity(tfidf_matrix[valid_indices], tfidf_matrix)
    
    # average similarity
    avg_sim = sim_matrix.mean(axis=0)

    # exclude input movies
    for idx in valid_indices:
        avg_sim[idx] = -1

    # top-N indices
    top_n = 10
    top_indices = avg_sim.argsort()[::-1][:top_n]

    # recommended movie titles
    recommended_titles = df['original_title'].iloc[top_indices]
    print(recommended_titles)
else:
    print(pd.Series([]))
```
